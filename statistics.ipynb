```python
import pandas as pd

data = pd.read_csv('data/big_cancer_data.csv')
data.describe()
```

```python
num_data = data.select_dtypes([int])
z_score = (num_data - num_data.mean()) / num_data.std()
z_score
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(2, 2, figsize=(15, 10))

sns.countplot(data, x="Smoking", hue="Level", ax=ax[0, 0])
sns.histplot(data, x="Age", hue="Level", ax=ax[0, 1], kde=True)
sns.violinplot(data, x="Gender", y="Smoking", ax=ax[1, 0])
sns.boxplot(data, x="Smoking", y="Age", hue="Level", ax=ax[1, 1])

fig.tight_layout()
plt.show()
```
